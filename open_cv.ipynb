```python
class_id = []


for word in df["image"]:
    if "meme" in word:
        class_id.append("meme")
    if "promo" in word:
        class_id.append("promo")
    if "landscape" in word and "wo" not in word:
        class_id.append("landscape_w")
    if "landscape_wo" in word:
        class_id.append("landscape_wo")
    if "animal" in word:
        class_id.append("animal")
    if "selfie" in word:
        class_id.append("selfie")
    if "object" in word:
        class_id.append("object")
    if "art" in word:
        class_id.append("art")
    if "body" in word:
        class_id.append("body")
    if "building" in word:
        class_id.append("building")
    if "candid" in word:
        class_id.append("candid")
    if "quote" in word:
        class_id.append("quote")
    if "urban" in word and "wo" not in word:
        class_id.append("urban_w")
    if "urban_wo" in word:
        class_id.append("urban_wo")
    if "food" in word:
        class_id.append("food")
    if "group" in word:
        class_id.append("group")
        
```

```python
# Calculate the histogram of features
im_features = np.zeros((len(arraylist), 100), "float32")
for i in range(len(arraylist)):
    words, distance = vq(des_list[i],vocabulary)
    for w in words:
        im_features[i][w] += 1
```
